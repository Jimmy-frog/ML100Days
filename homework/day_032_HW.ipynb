### Q1:計算標準常態分配，小於1的機率有多大?
![](Q1.png)

```python

# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
## Q1
#  提示: 用 cdf 來算
#pdf
mu=0
sigma=1
x=np.linspace(-4, 4, 500)



#cdf
cumsum_probs =  stats.norm.cdf(x,mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title(' cdf of Nomal(0,1)')
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()

sd = stats.norm.cdf(1,mu, sigma)
sd
```

### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?

```python
## Q2

big_1 = 1-stats.norm.cdf(1,mu, sigma)
less_1 = stats.norm.cdf(-1,mu, sigma)
less_1,big_1
```

### Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?
