```python

import matplotlib.pyplot as plt
import numpy as np
matplotlib.rc("font",family=['DFKai-SB',"Arial"])

```

```python
# 準備數據 ... 假設我要畫一個sin波 從0~180度
# 設定要畫的的x,y數據list....
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)

# 開始畫圖
plt.plot(x,y)
# 在這個指令之前，都還在做畫圖的動作 
# 這個指令算是 "秀圖"
plt.show() 

#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)

plt.show()

```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

#alpha點透明度;lineweight邊緣粗度;edgecolor邊緣顏色
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
#選擇象限（即分支）
T = np.arctan2(Y,X)

plt.scatter(X ,Y, alpha=0.5,linewidths=3, c=T)


# cm = plt.cm.get_cmap('RdYlBu')
# sc = plt.scatter(X, Y, color =(0,221,95))
# plt.scatter(X,Y, cmap=cm)

plt.show()
```
