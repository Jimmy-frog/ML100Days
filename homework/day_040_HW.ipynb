## KNN
## step 1 離散轉連續型資料

```python
data=data.copy() #複製一份資料給 data
#ValueError: could not convert string to float: 'Male', sklearn 中的 KNN 只能處理數值型態
#轉型 
#透過 preprocessing 轉換
from sklearn import preprocessing
#下面兩行程式，讓 Male=1, FeMale-0.
le = preprocessing.LabelEncoder()
data['Sex']=le.fit_transform(data['Sex'])
display(data)

df = pd.DataFrame({'sex':data['Sex'],'Pclass':data['Pclass'],'Age':data['Age']})
df
```

## Step 2 :計算資料點的倆倆距離

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.metrics.pairwise import nan_euclidean_distances

display(pd.DataFrame(nan_euclidean_distances(df)))
```

## Step3: 透過 KNN 進行補值

## Weight = uniform

```python
# KNN
# 設定 k 值
value_neighbors=1
from sklearn.impute import KNNImputer
#初始化: Initialize KNNImputer
imputer = KNNImputer(n_neighbors=value_neighbors)
df_filled = pd.DataFrame(imputer.fit_transform(df))
# display(data[missing_index.values==True])
# display(df_filled[missing_index.values==True])
display(df_filled)
```

```python
#K=3
value_neighbors=3
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=value_neighbors)
df_filled = pd.DataFrame(imputer.fit_transform(df))
#display(data[missing_index.values==True])
#display(df_filled[missing_index.values==True])
display(df_filled)
```

## weight = distance

```python
# KNN
# 設定 k 值
value_neighbors=1
from sklearn.impute import KNNImputer
#初始化: Initialize KNNImputer
imputer = KNNImputer(n_neighbors=value_neighbors,weights='distance')
df_filled = pd.DataFrame(imputer.fit_transform(df))
#display(data[missing_index.values==True])
#display(df_filled[missing_index.values==True])
display(df_filled)
'''
K=1，weight 比較看不出來
'''
```

```python
# KNN
# 設定 k 值
value_neighbors=3
from sklearn.impute import KNNImputer
#初始化: Initialize KNNImputer
imputer = KNNImputer(n_neighbors=value_neighbors,weights='distance')
df_filled = pd.DataFrame(imputer.fit_transform(df))
#display(data[missing_index.values==True])
#display(df_filled[missing_index.values==True])
display(df_filled)
```
